```python
import pandas as pd

# 1. 파일 불러오기
# 업로드된 파일을 확인한 결과, 이 파일은 탭이 아닌 콤마(,)로 구분되어 있습니다.
df = pd.read_csv('챗봇 데이터세트.csv', sep=',')

# 2. 일상 대화(label 0) 데이터만 추출
# 논문에서는 일상 인텐트를 구축하기 위해 label 0을 사용합니다.
df_daily = df[df['label'] == 0].copy()

# 3. 'Q' 열을 'utterance'로 변경하고 '일상' 인텐트 부여
# 사용자의 발화인 Q열만 사용합니다.
df_pre_chatbot = df_daily[['Q']].rename(columns={'Q': 'utterance'})
df_pre_chatbot['intent'] = '일상'

# 4. 결과 저장
df_pre_chatbot.to_csv('pre_chatbot.csv', index=False, encoding='utf-8-sig')

print(f"챗봇 일상 데이터 전처리 완료: {len(df_pre_chatbot)}건 저장")
```

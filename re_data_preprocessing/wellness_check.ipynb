```python
import pandas as pd

# 1. 파일 불러오기 (확인하고 싶은 파일명을 적으세요)
file_name = 'pre_wellness.csv'  # 정제 전 파일 확인 시
# file_name = 'cleaned_wellness.csv'  # 정제 후 파일 확인 시
df = pd.read_csv(file_name)

# 2. 전체 행이 완전히 똑같은 중복 데이터 개수 확인
duplicate_all = df.duplicated().sum()

# 3. 'utterance'(발화) 내용만 중복되는 데이터 개수 확인
# 인텐트는 다르더라도 문장이 같으면 중복으로 간주합니다.
duplicate_utterance = df.duplicated(subset=['utterance']).sum()

print(f"[{file_name} 데이터 중복 체크 결과]")
print(f"1. 전체 컬럼이 일치하는 중복 행 수: {duplicate_all}건")
print(f"2. 발화 내용(utterance)이 일치하는 중복 행 수: {duplicate_utterance}건")

# 4. 어떤 내용이 중복되었는지 상위 10개만 살펴보기
if duplicate_utterance > 0:
    print("\n--- 중복된 발화 내용 예시 (상위 10개) ---")
    # keep=False는 중복된 모든 행을 표시하라는 의미입니다.
    duplicate_samples = df[df.duplicated(subset=['utterance'], keep=False)].sort_values(by='utterance')
    print(duplicate_samples.head(10))
```

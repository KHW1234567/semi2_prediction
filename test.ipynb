## 데이터 합치기

```python
import pandas as pd
import numpy as np
import re
# 데이터 로드


# 말뭉치 training 데이터
fin_t=pd.read_excel(".\데이터셋\감성대화\Training_221115_add\원천데이터\감성대화말뭉치(최종데이터)_Training\감성대화말뭉치(최종데이터)_Training.xlsx", engine='openpyxl')

# 말뭉치 validation 데이터
fin_v=pd.read_excel(".\데이터셋\감성대화\Validation_221115_add\원천데이터\감성대화말뭉치(최종데이터)_Validation.xlsx", engine='openpyxl')


# 빈칸은 nan으로 결측치 처리
fin_t=fin_t.fillna("")
fin_v=fin_v.fillna("")

# 사람문장 1,2,3을 한 컬럼으로 합치기
fin_t["사람문장"]=fin_t["사람문장1"].astype(str)+fin_t["사람문장2"].astype(str)+fin_t["사람문장3"].astype(str)
fin_v["사람문장"]=fin_v["사람문장1"].astype(str)+fin_v["사람문장2"].astype(str)+fin_v["사람문장3"].astype(str)

# training과 validation 합치기
df_concat=pd.concat([fin_t,fin_v])
final_df=df_concat[["사람문장","감정_대분류"]]
final_df=final_df.rename({"감정_대분류":"감정"},axis=1)

#컬럼 내 데이터들의 공백(string) 제거
final_df["감정"]=final_df["감정"].apply(lambda x:x.strip())
final_df["감정"].value_counts()


final_df.head()
```

## 텍스트 정제

```python
# ─────────────────────────────────────────
# 2. 텍스트 정제
# ─────────────────────────────────────────
 
def clean_text(text: str) -> str:
    """특수문자·이모지 제거, 연속 공백 정리"""
    text = re.sub(r'[^\w\s가-힣.,!?]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
 
final_df["사람문장"] = final_df["사람문장"].apply(clean_text)
 
# 텍스트 길이 2자 미만 제거 (합치기 과정에서 생긴 빈 행 처리)
final_df = final_df[final_df["사람문장"].str.len() >= 2].reset_index(drop=True)
```

## 감정 라벨링

```python
# ─────────────────────────────────────────
# 3. 레이블 인코딩
# ─────────────────────────────────────────
 
EMOTION_MAP = {
    "불안": 0,
    "분노": 1,
    "상처": 2,
    "슬픔": 3,
    "당황": 4,
    "기쁨": 5,
}
 

# 알 수 없는 레이블 확인
unknown = final_df[~final_df["감정"].isin(EMOTION_MAP.keys())]["감정"].unique()
if len(unknown) > 0:
    print(f"⚠ 알 수 없는 레이블 (제거됨): {unknown}")
    final_df = final_df[final_df["감정"].isin(EMOTION_MAP.keys())].reset_index(drop=True)
 
# 6분류 레이블 (0~5)
final_df["label"] = final_df["감정"].map(EMOTION_MAP)
 
# 이진 레이블: 기쁨(5) → 0(긍정), 나머지 → 1(부정)
final_df["is_negative"] = (final_df["label"] != 5).astype(int) 
```

## 중복 제거

```python
# 중복된 행만 추출 (keep=False → 중복 관련 모든 행 표시)
duplicates = final_df[final_df.duplicated(subset=["사람문장"], keep=False)].copy()
duplicates = duplicates.sort_values("사람문장")

print(f"중복 관련 행 수: {len(duplicates)}")
print(f"중복된 고유 텍스트 수: {duplicates['사람문장'].nunique()}")
print()
duplicates.head(20)
```

```python
# ─────────────────────────────────────────
# 4. 중복 제거
# ─────────────────────────────────────────
 
before = len(final_df)
final_df = final_df.drop_duplicates(subset=["사람문장"]).reset_index(drop=True)
print(f"중복 제거: {before} → {len(final_df)} ({before - len(final_df)}건 제거)")
```

```python
# ─────────────────────────────────────────
# 5. 분포 확인
# ─────────────────────────────────────────
 
print("\n[6분류 감정 분포]")
print(final_df["감정"].value_counts())
 
print("\n[이진 분류 분포 (부정감정 탐지)]")
print(final_df["is_negative"].value_counts().rename({0: "긍정(기쁨)", 1: "부정"}))
 
# 클래스 불균형 경고
counts = final_df["label"].value_counts()
ratio = counts.max() / counts.min()
if ratio > 3:
    print(f"\n⚠ 클래스 불균형 감지 (max/min = {ratio:.1f}x) → class_weight='balanced' 권장")
```

```python
# ─────────────────────────────────────────
# 2. 텍스트 정제
# ─────────────────────────────────────────

import re
from konlpy.tag import Okt
okt = Okt()


def advanced_clean_text(text: str) -> str:
    """특수문자 제거 및 특허 기반 부정문 처리 규칙 적용"""
    # 한글, 영문, 숫자, 문장부호는 남기고 특수문자 제거
    text = re.sub(r'[^\w\s가-힣.,!?]', ' ', text)
    # 연속 공백 제거
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 형태소 분석 및 정규화 (Okt의 stem/norm 활용으로 오타 교정)
    # ex) "알쥐"->"알지" "슬프지"->"슬프다"
    tokens = okt.pos(text, stem=True, norm=True)
    
    # 특허 로직: 부정 어구(않다, 안, 못 등) 감지 시 앞 단어의 의미 반전
    # ex) "슬프지 않다"에서 "슬픔" 키워드만 추출하면 안됨
    # 슬프지 않은데 슬프다고 인식될 수 있기 때문
    # 뒤에 오는 부정문 "않다"를 감지해서 슬픔 감정을 반전시켜야한다
    negation_words = ['않다', '안', '못', '없다', '아니다']
    processed_tokens = []
    
    for i, (word, pos) in enumerate(tokens):
        if word in negation_words and processed_tokens:
            # 부정어가 오면 바로 앞의 감정 형용사/동사에 'NOT_' 접두사를 붙여 의미 보존
            processed_tokens[-1] = "NOT_" + processed_tokens[-1]
        if pos in ['Noun', 'Verb', 'Adjective']:
            processed_tokens.append(word)
            
    return " ".join(processed_tokens)

print("텍스트 정제 및 부정문 처리 중...")
final_df_cleaned["사람문장"] = final_df_cleaned["사람문장"].apply(advanced_clean_text)
# 텍스트 길이 2자 미만 제거 (합치기 과정에서 생긴 빈 행 처리)
final_df_cleaned = final_df_cleaned[final_df_cleaned["사람문장"].str.len() >= 2].reset_index(drop=True)
```

## 클래스별 가중치 적용

```python
# ─────────────────────────────────────────
# 3. 클래스 불균형 확인 및 class_weight 계산
# ─────────────────────────────────────────
print("[클래스 분포]")
label_counts = final_df["label"].value_counts().sort_index()
emotion_names = {0:"불안", 1:"분노", 2:"상처", 3:"슬픔", 4:"당황", 5:"기쁨"}
 
for label, count in label_counts.items():
    pct = count / len(final_df) * 100
    bar = "█" * int(pct / 2)
    print(f"  {label} {emotion_names[label]:3s} | {bar:25s} {count:6d}건 ({pct:.1f}%)")
 
# sklearn으로 class_weight 계산 (학습 시 손실함수에 전달)
classes = np.array(sorted(final_df["label"].unique()))
weights = compute_class_weight(
    class_weight='balanced',
    classes=classes,
    y=final_df["label"]
)
class_weight_dict = dict(zip(classes.tolist(), weights.tolist()))
 
print("\n[class_weight (학습 시 사용)]")
for label, w in class_weight_dict.items():
    print(f"  {label} {emotion_names[label]:3s}: {w:.4f}")
 
# 이진 분류용 pos_weight (부정 감정 비율)
neg_count = final_df["is_negative"].sum()
pos_count = len(final_df) - neg_count
binary_pos_weight = pos_count / neg_count
print(f"\n[이진 분류 pos_weight]: {binary_pos_weight:.4f}")
print(f"  긍정(기쁨): {pos_count}건 / 부정: {neg_count}건")
 
```

```python
import torch
from torch import nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np
from transformers import BertModel
from kobert_tokenizer import KoBERTTokenizer

# KoBERT 모델 사용
tokenizer = KoBERTTokenizer.from_pretrained('skt/kobert-base-v1')
bert_model = BertModel.from_pretrained('skt/kobert-base-v1')
```

```python
# 3. KoBERT 전용 Dataset 클래스 정의 
class KoBERTDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_len):
        self.tokenizer = tokenizer
        self.max_len = max_len
        # '사람문장' 컬럼의 텍스트와 'label' 컬럼의 감정 값을 리스트로 변환
        self.sentences = dataframe['사람문장'].tolist()
        self.labels = dataframe['label'].tolist()

    def __getitem__(self, i):
        # 텍스트 토큰화 및 인코딩 [1]
        # truncation=True와 padding='max_length'를 통해 길이를 통일합니다.
        encoding = self.tokenizer.encode_plus(
            self.sentences[i],
            add_special_tokens=True, # [CLS], [SEP] 추가 [4]
            max_length=self.max_len,
            return_token_type_ids=False,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(self.labels[i], dtype=torch.long)
        }

    def __len__(self):
        return len(self.labels)
```

```python
# 4. 데이터 로드 및 적용
train_df = pd.read_csv("model1_train.csv")
val_df = pd.read_csv("model1_val.csv")
test_df = pd.read_csv("model1_test.csv")

# 데이터셋 생성
train_dataset = KoBERTDataset(train_df, tokenizer, max_len)
val_dataset = KoBERTDataset(val_df, tokenizer, max_len)
test_dataset = KoBERTDataset(test_df, tokenizer, max_len)

# 5. DataLoader 생성 [2]
# 훈련 안정화를 위해 셔플(shuffle)을 적용합니다.
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=batch_size)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size)

print(f"데이터셋 준비 완료! 배치 개수: {len(train_dataloader)}")
```

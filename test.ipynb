```python
import pandas as pd
import numpy as np

# 데이터 로드
fin_t=pd.read_excel((r"C:\Users\asia\Desktop\ㅛㅔ미2\데이터셋\감성대화\Training_221115_add\원천데이터\감성대화말뭉치(최종데이터)_Training\감성대화말뭉치(최종데이터)_Training.xlsx"), engine='openpyxl')
fin_v=pd.read_excel((r"C:\Users\asia\Desktop\ㅛㅔ미2\데이터셋\감성대화\Validation_221115_add\원천데이터\감성대화말뭉치(최종데이터)_Validation.xlsx"), engine='openpyxl')

# 

# 결측치 처리
fin_t=fin_t.fillna("")
fin_v=fin_v.fillna("")

# 컬럼간 str 합치기
fin_t["사람문장"]=fin_t["사람문장1"].astype(str)+fin_t["사람문장2"].astype(str)+fin_t["사람문장3"].astype(str)
fin_v["사람문장"]=fin_v["사람문장1"].astype(str)+fin_v["사람문장2"].astype(str)+fin_v["사람문장3"].astype(str)

# 데이터프레임 합치기
df_concat=pd.concat([fin_t,fin_v])
final_df=df_concat[["사람문장","감정_대분류"]]
final_df=final_df.rename({"감정_대분류":"감정"},axis=1)

#컬럼 내 데이터들의 공백(string) 제거
final_df["감정"]=final_df["감정"].apply(lambda x:x.strip())
final_df["감정"].value_counts()

final_df.head()
```

```python

# 감정 컬럼에서 "상처"로 시작하는 문자열을 "슬픔"으로, "기쁨"으로 시작하는 문자열을 "행복"으로 변경
final_df["감정"]=np.where(final_df["감정"].str.match("상처"),"슬픔",final_df["감정"])
final_df["감정"]=np.where(final_df["감정"].str.match("기쁨"),"행복",final_df["감정"])
final_df=final_df.reset_index(drop=True)

# 컬럼명 변경
final_df.rename({"사람문장":"발화"},axis=1,inplace=True)
```

```python
# 연속적 대화 데이터셋 로드
con_dialogue=pd.read_excel(r'C:\Users\asia\Desktop\ㅛㅔ미2\데이터셋\한국어_연속적_대화_데이터셋.xlsx')
con_dialogue.head()
```

```python
# 첫행을 컬럼으로 바꾸기
new_columns=con_dialogue.iloc[0]
con_dialogue=con_dialogue.rename(columns=new_columns).iloc[:,1:3]
con_dialogue.drop(index=0,axis=0,inplace=True)
con_dialogue.head()
```

```python
con_dialogue["감정"].value_counts()

# 오타 수정하기
# 분ㄴ -> 분노
# 분 -> 분노
# ㅈ중립 -> 중립
# 중림 -> 중립
# ㄴ중립 -> 중립
# 줄 -> 중립
# 'ㅍ' -> ????

correction_dict = {
    '분ㄴ': '분노',
    '분': '분노',
    '중림': '중립',
    'ㅈ중립': '중립',
    'ㄴ중립': '중립',
    '줄': '중립'
}

con_dialogue['감정'] = con_dialogue['감정'].replace(correction_dict)

#%%
con_dialogue["감정"].value_counts()
```

```python
con_dialogue=con_dialogue[con_dialogue["감정"].isin(["중립","놀람","분노","슬픔","행복","혐오","공포"])]
# 감정 Label 일정하게 맞춰주기 (감성대화 말뭉치와 통합을 위해)

con_dialogue["감정"]=np.where(con_dialogue["감정"].str.match("놀람"),"당황",con_dialogue["감정"])
con_dialogue["감정"]=np.where(con_dialogue["감정"].str.match("공포"),"불안",con_dialogue["감정"])

con_dialogue["감정"].value_counts()
```

```python
# 데이터 합치기

real_final_df=pd.concat([final_df,con_dialogue])
real_final_df.reset_index(drop=True,inplace=True)
real_final_df.to_csv("real_final_df.csv",index=False,encoding="utf-8-sig")
```

```python
# 사람문장과 시스템문장 분리하기
shuffled_train = fin_t.sample(frac=1).reset_index(drop=True)
shuffled_test = fin_v.sample(frac=1).reset_index(drop=True)


shuffled_train.head()

rows = []

# 각 행에 대해 사람문장과 시스템문장을 분리
for idx, row in shuffled_train.iterrows():
    # 사람문장1에 대한 시스템문장1
    if pd.notna(row['사람문장1']) and pd.notna(row['시스템문장1']):
        rows.append({
            '연령': row['연령'],
            '성별': row['성별'],
            '상황키워드': row['상황키워드'],
            '신체질환': row['신체질환'],
            '감정_대분류': row['감정_대분류'],
            '감정_소분류': row['감정_소분류'],
            '사람문장': row['사람문장1'],
            '시스템문장': row['시스템문장1']
        })
    
    # 사람문장2에 대한 시스템문장2
    if pd.notna(row['사람문장2']) and pd.notna(row['시스템문장2']):
        rows.append({
            '연령': row['연령'],
            '성별': row['성별'],
            '상황키워드': row['상황키워드'],
            '신체질환': row['신체질환'],
            '감정_대분류': row['감정_대분류'],
            '감정_소분류': row['감정_소분류'],
            '사람문장': row['사람문장2'],
            '시스템문장': row['시스템문장2']
        })

    # 사람문장3에 대한 시스템문장3 (NaN이 아닌 경우에만 추가)
    if pd.notna(row['사람문장3']) and pd.notna(row['시스템문장3']):
        rows.append({
            '연령': row['연령'],
            '성별': row['성별'],
            '상황키워드': row['상황키워드'],
            '신체질환': row['신체질환'],
            '감정_대분류': row['감정_대분류'],
            '감정_소분류': row['감정_소분류'],
            '사람문장': row['사람문장3'],
            '시스템문장': row['시스템문장3']
        })

# 새로운 데이터 프레임 생성
train = pd.DataFrame(rows)

train.head()
```
